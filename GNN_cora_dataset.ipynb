```python
import torch
from torch_geometric.datasets import TUDataset

dataset = TUDataset(root = '/tmp/ENZYMES' , name = 'ENZYMES')
```

```python
len(dataset)
```

```python
dataset[0]
```

```python
dataset.num_classes
```

```python
from torch_geometric.utils import to_networkx
import matplotlib.pyplot as plt
import networkx as nx

for i in range(5):
  data = dataset[i]        # getting a single graph
  graph = to_networkx(data)
  plt.figure()         # if we don't use plt it is making a single graph
  nx.draw(graph)
  plt.title(data.y.item()) 
  plt.show()

```

```python
nx.write_gexf(graph, 'cora_ex.gexf')       # it is used to save the graph as gexf file and can be viewed by respective app
nx.write_graphml(graph, 'cora_ex.graphml')
```
