# NETWORK X 
We use graphs because it is easy to show the relationship between different nodes representing different entities.
Example: Facebook uses undirected graph while twitter uses directed graph.


**refer networkx official library for more details**


```python
import networkx as nx   
G = nx.Graph()     # initialises the graph
# G = nx.DiGraph()     # for directed graph 
nx.draw(G)     # prints the graph
```

```python
G.add_node(5)    # draws a single node
G.add_nodes_from([3,4,6])     # draws multiple nodes
G.add_edge(5,9)      # adds an edge 
G.add_edges_from([(4,8), (7,9)])    # for multiple edges
nx.draw(G) 
```

```python
G.number_of_nodes()
G.number_of_edges()
G.edges()
G.nodes()
```

```python
for line in nx.generate_graphml(G):     # file format of graph
  print(line)
```

```python
# G.clear()      # clears the graph 
```

```python
A = nx.to_numpy_array(G)
print(A)
```
